# Logistic Regression Classifier Project

### Step 1 : Load Data Set

```python
import numpy as np
import pandas as pd
```

```python
data_set = pd.read_excel('student-data.xlsx')
```

### Step 2 : Analyze Data Set

```python
data_set.info()
```

```python
data_set.describe()
```

```python
data_set['Result'].value_counts()
```

### Step 3 : Handle Missing Values

```python
data_set['Hours_Studied'] = data_set['Hours_Studied'].fillna(data_set['Hours_Studied'].median())

data_set['Attendance_Percentage'] = data_set['Attendance_Percentage'].fillna(data_set['Attendance_Percentage'].median())

data_set['Previous_Score'] = data_set['Previous_Score'].fillna(data_set['Previous_Score'].median())
```

```python
data_set.isnull().sum()
```

### Step 4: Encode Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
data_set['Result'] = encoder.fit_transform(data_set['Result'])
```

```python
data_set.head()
```

### Step 5: Select Input and Output Features

```python
x = data_set[['Hours_Studied', 'Attendance_Percentage', 'Previous_Score']]
y = data_set[['Result']]
```

### Step 6: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
print(x[0])
```

### Step 7: Split into Train and Test Sets

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.3,
    random_state=42
)
```

### Step 8: Train Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver = 'liblinear', penalty='l1')
```

```python
model.fit(x_train, y_train)
```

### Step 9 : Testing the model

```python
prediction = model.predict(x_test)
```

### Step 10 : Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(prediction, y_test)
```

```python
print(score)
```

### Step 11 : Save the Model

```python
import pickle
with open('model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
