```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
x, y_true = make_blobs(n_samples=500, n_features=2, centers=4, cluster_std=0.60, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
data = pd.DataFrame(data=x, columns=['feature-1', 'feature-2'])
```

```python
from sklearn.cluster import KMeans
wcss = []
k_range = range(1,11)

for i in k_range:
    kmean = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmean.fit(x)
    wcss.append(kmean.inertia_)
```

```python
wcss = np.round(wcss,0)
```

```python
plt.plot(k_range, wcss, marker='o')
```

init is a random intialization trap ( k-mean++ ) solve it to choose long distance centriods from each other , you can use random init as well.

```python
kmean = KMeans(n_clusters=4, init='k-means++', random_state=42)
labels = kmean.fit_predict(data)
```

```python
data['labels'] = labels
```

```python
sns.scatterplot(x=data['feature-1'], y=data['feature-2'], hue=data['labels'])
```
