# Support Vector Mechine Regression

### Setp 1 : Load Data Set

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
x = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.DataFrame(data.target, columns=["MedHouseValue"])

data_set = pd.concat([x, y], axis=1)
```

### Step 3 : Handle Missing Values

```python
data_set.fillna(data_set.median(), inplace=True)
```

```python
data_set.isnull().sum()
```

```python
data_set.info()
```

```python
data_set.describe()
```

### Setp 4 : Split the Data

```python
features = data_set
output = data_set["MedHouseValue"]
features = features.drop("MedHouseValue", axis=1)
```

```python
features
```

```python
output
```

### Step 5 : Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    features, output,
    test_size=0.2,
    random_state=42
    )
```

### Step 6 : Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

### Step 7 : Train the Model

```python
from sklearn.svm import SVR
model = SVR()
```

```python
model.fit(x_train, y_train)
```

### Step 8 : Testing the Model

```python
prediction = model.predict(x_test)
```

### Step 9 : Evaluating the Model

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import r2_score
```

```python
mae = mean_absolute_error(y_test, prediction)
mse = mean_squared_error(y_test, prediction)
rmse = root_mean_squared_error(y_test, prediction)
r2score = r2_score(y_test, prediction)
```

```python
print(f"MAE : {mae}\nMSE : {mse}\nRMSE : {rmse}\nR2-SCORE : {r2score}")
```

### Step 10 : Save the Model

```python
import pickle
with open("svr_model.pkl", "wb") as f:
    pickle.dump(model, f)
```
